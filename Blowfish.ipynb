```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('Blowfish_performance.csv')

# make new file name based on File Type Column
df['File Type'] = df['File Type'].str.replace('.', '')
# make a new file for each file type
for file_type in df['File Type'].unique():
    df[df['File Type'] == file_type].to_csv('Blowfish_performance_analysis_' + file_type + '.csv', index=False)
```

```python
df = pd.read_csv('Blowfish_performance_analysis_txt.csv')
sns.pairplot(df)
plt.suptitle('Text Files')
summary_txt = df.describe()
# print(summary)
print(summary_txt.loc[["mean","std","min","max"]])
```

```python
df = pd.read_csv('Blowfish_performance_analysis_mp3.csv')
sns.pairplot(df)
plt.suptitle('MP3 Files')
summary_mp3 = df.describe()
# print(summary)

print(summary_mp3.loc[["mean","std","min","max"]])
# type(summary)
```

```python
df = pd.read_csv('Blowfish_performance_analysis_mp4.csv')
sns.pairplot(df)
plt.suptitle('MP4 Files')
summary_mp4 = df.describe()
# print(summary)
print(summary_mp4.loc[["mean","std","min","max"]])
```

```python
df = pd.read_csv('Blowfish_performance_analysis_png.csv')
sns.pairplot(df)
plt.suptitle('PNG Files')
summary_png = df.describe()
# print(summary)
print(summary_png.loc[["mean","std","min","max"]])
```

```python
summary_mp3['File Type'] = 'mp3'
summary_mp4['File Type'] = 'mp4'
summary_png['File Type'] = 'png'
summary_txt['File Type'] = 'txt'
print(summary_mp3)
```

```python
sum_final = pd.concat([summary_txt, summary_mp3, summary_mp4, summary_png])
sum_final
```

```python
sum_final['Statistics'] = ['Count', 'Mean', 'Std Dev', 'Min', '25%', 'Median', '75%', 'Max'] * 4
```

```python
sum_final = sum_final.set_index(['File Type', 'Statistics'])
```
